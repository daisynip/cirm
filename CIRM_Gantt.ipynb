# Gantt Chart

### Prepare the Environment

##### Load Data

```python
import pandas as pd
```

```python
xls = pd.ExcelFile(r"https://www.cirm.ca.gov/grants/export/xls")
```

```python
sheet=xls.parse(0)
```

```python
sheet
```

##### Data Wrangling

```python
from plotly.figure_factory import create_gantt
```

```python
df_gantt_pre = sheet[["Institution", "Start Date", "End Date", "Award Status"]]
df_gantt_pre["Task Number"] = df_gantt_pre.index
df_gantt_pre["Task Number"] = df_gantt_pre["Task Number"].astype(str) + " " + df_gantt_pre["Institution"]
df_gantt_pre.rename(columns={"Task Number": "Task", "Start Date": "Start", "End Date": "Finish", "Institution": "Resource"}, inplace=True)
df_gantt_pre
```

```python
df_gantt = df_gantt_pre.dropna()
df_gantt = df_gantt[["Task", "Start", "Finish", "Award Status", "Resource"]]
df_gantt["Start"] = df_gantt["Start"].astype("datetime64[ns]")
df_gantt["Start"] = pd.to_datetime(df_gantt["Start"], format="$y$m$d")
df_gantt["Finish"] = df_gantt["Finish"].astype("datetime64[ns]")
df_gantt["Finish"] = pd.to_datetime(df_gantt["Finish"], format="$y$m$d")
df_gantt
```

```python
df_gantt_2022 = df_gantt[(df_gantt["Start"] >= "2022-01-01") & (df_gantt["Start"] <= "2022-12-31")]
df_gantt_2022
```

### Visualization

```python
from datetime import date

fig = create_gantt(df_gantt_2022, showgrid_x=True, showgrid_y=True, height=1000)
today = date.today()
fig.update_layout(shapes=[
    dict(
      type='line',
      yref='paper', y0=0, y1=1,
      xref='x', x0=today, x1=today
    )
])
fig.show()
```

```python
import plotly.express as px
```

```python
fig = px.timeline(df_gantt_2022, x_start="Start", x_end="Finish", y="Task", color="Resource", height=1000)
fig.update_yaxes(autorange="reversed")
fig.update_layout(shapes=[
    dict(
      type='line',
      yref='paper', y0=0, y1=1,
      xref='x', x0=today, x1=today
    )
])
fig.show()
```
